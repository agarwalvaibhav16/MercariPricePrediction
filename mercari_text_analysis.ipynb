```python
def removespecialCharactersfromTokens(token):
    for i,z in enumerate(token):  
        token[i]=z.translate ({ord(c): " " for c in "!@#$%^&*()[]{};:,./<>?\|`~-=_+"}).strip(' ')
    return token
  
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def removeStopwords(token):
        return [t for t in token if t.lower() not in stopwords]
    
df_miss_50000['tokens']=df_miss_50000['tokens'].apply(lambda x:removeStopwords(x))


```

```python
def removingemptyTokens(token):
    return [t for t in token if len(t)>0]
    
df_miss_50000['tokens']=df_miss_50000['tokens'].apply(lambda x:removingemptyTokens(x))
```

```python
def getSize(token):
    indices=[ind for ind,w in enumerate(token) if re.search('^[Ss][i][z][e]$',w)]
#   print(indices)
    sizes=[token[index+1] for index in indices if index<len(token)-1]
    return sizes

df_miss_50000['Size']=df_miss_50000['tokens'].apply(lambda x:getSize(x))
```
